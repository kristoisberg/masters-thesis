```python
root_dir = ".."
analyser_path = "../../statement-counter/build/libs/statement-counter-1.0-SNAPSHOT-all.jar"
```

```python
import pandas as pd
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
projects_df = pd.read_csv(f"{root_dir}/datasets/final-repositories.csv")
projects_df = projects_df.drop(columns=["response_code", "result_count", "response_code2", "result_count2"])
projects_df
```

```python
def get_project_dir(project: str):
    return f"{root_dir}/repositories/{project.replace("/", "_")}"
```

```python
import os

def run_analyser(project_name):
    project_path = get_project_dir(project_name)
    # Construct the command string
    command = f"java -jar {analyser_path} {project_path}"

    try:
        # Open a pipe to the command
        with os.popen(command) as stream:
            output = stream.read().strip()
            # Return integer if output exists, otherwise None
            return int(output) if output else None
    except Exception as e:
        print(f"Error processing {project_name}: {e}")
        return None
```

```python
projects_df["count"] = projects_df["name"].apply(run_analyser)
```
