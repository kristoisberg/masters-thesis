```python
import pandas as pd
```

```python
ground_truth_df = pd.read_csv('../datasets/ground-truth.csv')
ground_truth_df
```

```python
# Each project contained at least one antipattern so this is valid
total_project_count = ground_truth_df["Project"].unique().size
total_project_count
```

```python
result_df = pd.DataFrame(ground_truth_df["Antipattern"].unique(), columns=["Antipattern"])
result_df = result_df.assign(
    **{
        "Instance count": 0,
        "Containing project count": 0, 
        "Instances per any project": 0.0, 
        "Instances per containing project": 0.0
    }
)
result_df
```

```python
# Number of total occurrences
for index, row in result_df.iterrows():
    count = ground_truth_df[ground_truth_df["Antipattern"] == row["Antipattern"]].shape[0]
    result_df.loc[index, "Instance count"] = count
```

```python
# Number of antipatterns instances per project (including projects that do not contain the antipattern)
for index, row in result_df.iterrows():
    count = ground_truth_df[ground_truth_df["Antipattern"] == row["Antipattern"]].shape[0] / total_project_count
    result_df.loc[index, "Instances per any project"] = count
```

```python
# Number of projects containing each antipattern
for index, row in result_df.iterrows():
    count = ground_truth_df[ground_truth_df["Antipattern"] == row["Antipattern"]]["Project"].unique().size
    result_df.loc[index, "Containing project count"] = count
```

```python
# Number of antipatterns instances per project (only including projects that contain the antipattern)
for index, row in result_df.iterrows():
    df_antipattern = ground_truth_df[ground_truth_df["Antipattern"] == row["Antipattern"]]
    project_count = df_antipattern["Project"].unique().size
    result_df.loc[index, "Instances per containing project"] = df_antipattern.shape[0] / project_count
```

```python
result_df.sort_values(by=["Instance count", "Containing project count"], ascending=False, ignore_index=True)
```

```python
included_antipatterns_df = result_df[result_df["Instance count"] > 25]
included_antipatterns_df = included_antipatterns_df[included_antipatterns_df["Containing project count"] > 3]
included_antipatterns_df.sort_values(by=["Instance count", "Containing project count"], ascending=False, ignore_index=True)
```
