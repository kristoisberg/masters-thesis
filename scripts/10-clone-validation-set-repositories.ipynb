```python
# Used to clone only the validation set in storage-constrained environments
```

```python
import pandas as pd
import os
```

```python
df = pd.read_csv('../datasets/validation-set.csv')
df
```

```python
projects = df["Project"].unique().tolist()
projects
```

```python
for project in projects:
    source_url = f"https://github.com/{project}.git"
    target_dir = f"../repositories/{project.replace('/', '_')}"
    
    if os.path.isdir(target_dir) == False:
        os.system(f"git clone {source_url} {target_dir} --depth 1")
```
