```python
import requests_async
import os
import time
import pandas as pd
```

```python
df = pd.read_csv('../datasets/all-repositories.csv')
df = df.assign(response_code = None, result_count = None)
df
```

```python
async def request(repository):
    return await requests_async.get(
        'https://api.github.com/search/code', 
        params = {'q': f"repo:{row['name']} language:java \"A class modelling foreign key relationships and constraints of tables\""}, 
        headers = {'Authorization': 'Bearer ' + os.getenv('GITHUB_TOKEN'), 'X-GitHub-Api-Version': '2022-11-28'}
    )

for index, row in df.iterrows():
    if row['response_code'] == None or row['response_code'] == 403:
        response = await request(row)
        df.loc[index, 'response_code'] = response.status_code
        df.loc[index, 'result_count'] = response.json()['total_count']
        print(row['name'], response.status_code, response.json()['total_count'])
        time.sleep(6.5)
```

```python
print(df['response_code'].value_counts())
print(df['result_count'].value_counts())
df
```

```python
df_valid = df[df['result_count'] > 0]
df_valid
```
