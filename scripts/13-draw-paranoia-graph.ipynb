```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
models = [
    ["GPT-5.2", 0.8537, 0.9225],
    ["GLM-5", 0.8837, 0.8933],
    ["Opus 4.5", 0.8836, 0.8898],
    ["gpt-oss-120B", 0.8262, 0.8692]
]

models_df = pd.DataFrame(models, columns=["Model", "precision", "recall"])
models_df
```

```python
model_colors = {
    'GPT-5.2': "#10A37F",
    'Opus 4.5': "#FF9436",
    'GLM-5':  '#002366',
    'gpt-oss-120B': '#6200EE'
}
```

```python
plt.figure(figsize=(10, 10))

sns.set_theme(style="white")

sns.scatterplot(
    data=models_df,
    x='precision',
    y='recall',
    hue='Model',
    palette=model_colors,
    s=100
)

plt.xlim(0, 1.0)
plt.ylim(0, 1.0)
plt.xlabel("Precision", fontsize=12, labelpad=10)
plt.ylabel("Recall", fontsize=12, labelpad=10)
plt.grid(True)

plt.show()
```
