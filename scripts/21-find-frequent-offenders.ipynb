```python
root_dir = ".."
```

```python
results_df = pd.read_csv(f"{root_dir}/datasets/analysis-results.csv")
results_df
```

```python
implicit_columns_df = results_df[results_df['Antipattern'] == 'Implicit Columns'].copy()
implicit_columns_df
```

```python
import re

def categorize_fragment(code):
    if 'fields()' in code:
        return 'fields()'
    if 'asterisk()' in code:
        return 'asterisk()'

    if 'returning()' in code:
        return 'returning()'
    if 'returningResult()' in code:
        return 'returningResult()'
    if re.search(r'\.returning\(\s*([^.\s,)]+)\s*[,)]', code) or re.search(r'\.returning\(Tables\.\s*([^.\s,)]+)\s*[,)]', code):
        return 'returning(TABLE'
    if re.search(r'\.returningResult\(\s*([^.\s,)]+)\s*[,)]', code) or re.search(r'\.returningResult\(Tables\.\s*([^.\s,)]+)\s*[,)]', code) or 'returningResult(Wallet.WALLET)' in code or 'returningResult(UsersTable.USERS_TABLE)' in code:
        return 'returningResult(TABLE'

    if 'selectFrom' in code:
        return 'selectFrom(TABLE)'
    if 'select()' in code:
        return 'select().from(TABLE)'
    if 'selectDistinct()' in code:
        return 'selectDistinct().from(TABLE)'
    if re.search(r'\.select\(\s*([^.\s\)]+)\s*', code) or re.search(r'\.select\(Tables\.\s*([^.\s\)]+)\s*', code) or 'select(Degente.DEGENTE)' in code:
        return 'select(TABLE).from(TABLE)'
    if re.search(r'\.select\(\s*[^.\s\)]+(?:\s*,\s*[^.\s\)]+)+\s*', code) or re.search(r'\.select\(Tables\.\s*[^.\s\)]+(?:\s*,Tables\.\s*[^.\s\)]+)+\s*', code):
        return 'select(TABLE1, TABLE2)'

    if 'dao.fetchoneby' in code.lower() or 'dao::fetchoneby' in code.lower() or 'dao().fetchoneby' in code.lower():
        return 'dao.fetchOneBy'
    if 'dao.fetchoptionalby' in code.lower() or 'dao::fetchoptionalby' in code.lower() or 'dao().fetchoptionalby' in code.lower():
        return 'dao.fetchOptionalBy'
    if 'dao.fetchby' in code.lower() or 'dao::fetchby' in code.lower() or 'dao().fetchby' in code.lower():
            return 'dao.fetchBy'
    if 'dao.fetchrangeof' in code.lower() or 'dao::fetchrangeof' in code.lower() or 'dao().fetchrangeof' in code.lower():
            return 'dao.fetchRangeOf'
    if 'dao.fetch' in code.lower() or 'dao::fetch' in code.lower() or 'dao().fetch' in code.lower():
            return 'dao.fetch'
    if 'dao.findall' in code.lower() or 'dao::findall' in code.lower() or 'dao().findall' in code.lower():
            return 'dao.findAll'
    if 'dao.findbyid' in code.lower() or 'dao::findbyid' in code.lower() or 'dao().findbyid' in code.lower():
            return 'dao.findById'
    if 'dao.findonebyid' in code.lower() or 'dao::findonebyid' in code.lower() or 'dao().findonebyid' in code.lower():
            return 'dao.findOneById'
    if 'dao.findoptionalbyid' in code.lower() or 'dao::findoptionalbyid' in code.lower() or 'dao().findoptionalbyid' in code.lower():
            return 'dao.findOptionalById'
    if 'dao.findmanyby' in code.lower() or 'dao::findmanyby' in code.lower() or 'dao().findmanyby' in code.lower():
            return 'dao.findManyBy'
    if 'dao.findoneby' in code.lower() or 'dao::findoneby' in code.lower() or 'dao().findoneby' in code.lower():
            return 'dao.findOneBy'
    
    if re.search(r'\.fetchOne\(\s*([^.\s,)]+)\s*[,)]', code) or re.search(r'\.fetchOne\(Tables\.\s*([^.\s,)]+)\s*[,)]', code) or 'fetchOne(Accounts.ACCOUNTS' in code:
        return 'fetchOne(TABLE'
    if re.search(r'\.fetchSingle\(\s*([^.\s,)]+)\s*[,)]', code) or re.search(r'\.fetchSingle\(Tables\.\s*([^.\s,)]+)\s*[,)]', code):
        return 'fetchSingle(TABLE'
    if re.search(r'\.fetch\(\s*([^.\s,)]+)\s*[,)]', code) or re.search(r'\.fetch\(Tables\.\s*([^.\s,)]+)\s*[,)]', code):
        return 'fetch(TABLE'
    if re.search(r'\.fetchOptional\(\s*([^.\s,)]+)\s*[,)]', code) or re.search(r'\.fetchOptional\(Tables\.\s*([^.\s,)]+)\s*[,)]', code):
        return 'fetchOptional(TABLE'
    
    if '*' in code:
        return 'raw SELECT *'

    # zero-ws/zero-ecotope abstractions
    if 'fetchAsync' in code:
        return 'ZERO fetchAsync'
    if 'fetchOneAsync' in code:
        return 'ZERO fetchOneAsync'
    if 'fetchByIdAsync' in code:
        return 'ZERO fetchByIdAsync'
    if 'fetchInAsync' in code:
        return 'ZERO fetchInAsync'
    if 'fetchAndAsync' in code:
        return 'ZERO fetchAndAsync'
    if 'fetchAllAsync' in code:
        return 'ZERO fetchAllAsync'
    if 'fetchJAsync' in code:
        return 'ZERO fetchJAsync'
    if 'fetchJOneAsync' in code:
        return 'ZERO fetchJOneAsync'
    if 'fetchJByIdAsync' in code:
        return 'ZERO fetchJByIdAsync'
    if 'fetchJInAsync' in code:
        return 'ZERO fetchJInAsync'
    if 'fetchJAndAsync' in code:
        return 'ZERO fetchJAndAsync'

    return "Other/Specific Method"

# Apply the logic
implicit_columns_df['clean_code'] = implicit_columns_df['Code fragment'].str.replace(r'\s|\\n', '', regex=True)
implicit_columns_df['method_type'] = implicit_columns_df['clean_code'].apply(categorize_fragment)
```

```python
# Get the counts of each method manifestation
top_methods = implicit_columns_df['method_type'].value_counts()

print("Common Methods associated with Implicit Columns:")
print(top_methods)

# To see which projects are the biggest offenders for specific methods:
pivot = implicit_columns_df.groupby(['project_name', 'method_type']).size().unstack(fill_value=0)
```

```python
poor_mans_search_engine_df = results_df[results_df['Antipattern'] == 'Poor Man\'s Search Engine'].copy()
poor_mans_search_engine_df
```

```python
import re

def categorize_fragment(code):
    if 'containsIgnoreCase(' in code:
        return 'containsIgnoreCase'
    if 'contains(' in code:
        return 'contains'
    if 'endsWithIgnoreCase(' in code:
        return 'endsWithIgnoreCase'
    if 'endsWith(' in code:
        return 'endsWith'
    if 'notContainsIgnoreCase(' in code:
        return 'notContainsIgnoreCase'
    if 'notContains(' in code:
        return 'notContains'
    if 'likeIgnoreCase(' in code:
        return 'likeIgnoreCase'
    if 'like(' in code:
        return 'like'
    if 'notLike(' in code:
        return 'notLike('
    if 'likeRegex' in code:
        return 'likeRegex'
    if 'notLikeRegex' in code:
        return 'notLikeRegex'
    if 'similarTo' in code:
        return 'similarTo'
    if 'notSimilarTo' in code:
        return 'notSimilarTo'
    if 'ilike\'%' in code.lower() or 'ilike\\"%' in code.lower() or 'ilike?' in code.lower():
        return 'Plain SQL ILIKE'
    if 'like\'%' in code.lower() or 'like\\"%' in code.lower() or 'like?' in code.lower() or 'LIKElower("+"\'%"' in code:
        return 'Plain SQL LIKE'
    return "Other/Specific Method"

# Apply the logic
poor_mans_search_engine_df['clean_code'] = poor_mans_search_engine_df['Code fragment'].str.replace(r'\s|\\n', '', regex=True)
poor_mans_search_engine_df['method_type'] = poor_mans_search_engine_df['clean_code'].apply(categorize_fragment)
```

```python
# Get the counts of each method manifestation
top_methods = poor_mans_search_engine_df['method_type'].value_counts()

print("Common Methods associated with Poor Man's Search Engine:")
print(top_methods)

# To see which projects are the biggest offenders for specific methods:
pivot = poor_mans_search_engine_df.groupby(['project_name', 'method_type']).size().unstack(fill_value=0)
```
