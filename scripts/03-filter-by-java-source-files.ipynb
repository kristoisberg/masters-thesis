```python
import requests_async
import os
import time
import pandas as pd
```

```python
df = pd.read_csv('../datasets/repositories-with-generated-classes.csv')
df = df.assign(response_code2 = None, result_count2 = None)
df
```

```python
async def request(repository):
    return await requests_async.get(
        'https://api.github.com/search/code', 
        params = {'q': f"repo:{row['name']} language:java \"import org.jooq.DSLContext\""}, 
        headers = {'Authorization': 'Bearer ' + os.getenv('GITHUB_TOKEN'), 'X-GitHub-Api-Version': '2022-11-28'}
    )

for index, row in df.iterrows():
    if row['response_code2'] == None or row['response_code2'] == 403:
        response = await request(row)
        df.loc[index, 'response_code2'] = response.status_code
        df.loc[index, 'result_count2'] = response.json()['total_count']
        print(row['name'], response.status_code, response.json()['total_count'])
        time.sleep(6.5)
```

```python
print(df['response_code2'].value_counts())
print(df['result_count2'].value_counts())
df
```

```python
df_valid = df[df['result_count2'] > 0]
df_valid
```

```python
df_valid.to_csv('../datasets/repositories-with-java-source-files.csv', index=False)
```
