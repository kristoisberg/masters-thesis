```python
import pandas as pd
import os
```

```python
df = pd.read_csv('../datasets/repositories-with-java-source-files.csv')
df
```

```python
for index, row in df.iterrows():
    source_url = f"https://github.com/{row['name']}.git"
    target_dir = f"../repositories/{row['name'].replace('/', '_')}"
    
    if os.path.isdir(target_dir) == False:
        os.system(f"git clone {source_url} {target_dir} --depth 1")
```
