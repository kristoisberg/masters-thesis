```python
reannotated_files_df = pd.read_csv('../datasets/files-to-reannotate.csv')
reannotated_files_df
```

```python
new_annotations_df = pd.read_csv('../datasets/reannotated-ground-truth.csv')
new_annotations_df
```

```python
original_annotations_df = pd.read_csv('../datasets/ground-truth.csv')
original_annotations_df
```

```python
original_annotations_df = original_annotations_df.merge(
    reannotated_files_df[["Project", "File"]], on=["Project", "File"], how="inner"
).sort_values(by=["Project", "File"], ignore_index=True)
original_annotations_df
```

```python
import pandas as pd
import numpy as np

# --- 1. The Helper Functions (As provided) ---

def iou_1d_inclusive(interval_a, interval_b):
    start_inter = max(interval_a[0], interval_b[0])
    end_inter = min(interval_a[1], interval_b[1])
    inter_len = max(0, end_inter - start_inter + 1)
    
    len_a = interval_a[1] - interval_a[0] + 1
    len_b = interval_b[1] - interval_b[0] + 1
    union_len = len_a + len_b - inter_len
    
    return inter_len / union_len if union_len > 0 else 0.0

# --- 2. Confusion Matrix Construction ---

# Define the categories
antipatterns = sorted(list(set(original_annotations_df['Antipattern']).union(set(new_annotations_df['Antipattern']))))
categories = antipatterns + ["Background"]
cat_to_idx = {cat: i for i, cat in enumerate(categories)}

# Initialize confusion matrix (N x N)
# Rows: Original (Annotator A), Cols: New (Annotator B)
k = len(categories)
conf_matrix = np.zeros((k, k), dtype=int)

# Get unique file/project pairs to iterate through
all_files = pd.concat([
    original_annotations_df[['Project', 'File']], 
    new_annotations_df[['Project', 'File']]
]).drop_duplicates()

# Keep track of which rows were "matched" to handle the Background category
matched_indices_new = set()

for _, row in all_files.iterrows():
    proj, filename = row['Project'], row['File']
    
    # Filter for current file
    df_a = original_annotations_df[(original_annotations_df['Project'] == proj) & (original_annotations_df['File'] == filename)]
    df_b = new_annotations_df[(new_annotations_df['Project'] == proj) & (new_annotations_df['File'] == filename)]
    
    # Convert to list of dicts for easier tracking
    list_a = df_a.to_dict('records')
    list_b = df_b.to_dict('records')
    
    # Track which items in B have been matched to an item in A
    matched_in_b_for_this_file = [False] * len(list_b)
    
    # 1. Initialize match tracker for B
    matched_in_b_indices = set()
    
    # 2. FIRST PASS: Iterate through A to find matches in B
    for idx_a, item_a in df_a.iterrows():
        span_a = (item_a['Line from'], item_a['Line to'])
        type_a = item_a['Antipattern']
        found_match = False
        
        for idx_b, item_b in df_b.iterrows():
            # Skip if this annotation in B is already matched
            if idx_b in matched_in_b_indices:
                continue
                
            span_b = (item_b['Line from'], item_b['Line to'])
            if iou_1d_inclusive(span_a, span_b) >= 0.5:
                # We found a spatial match! 
                # (Note: In Kappa, we record what B actually labeled it)
                type_b = item_b['Antipattern']
                conf_matrix[cat_to_idx[type_a]][cat_to_idx[type_b]] += 1
                matched_in_b_indices.add(idx_b)
                found_match = True
                break
                
        if not found_match:
            # Annotator A found it, B did not (Background)
            conf_matrix[cat_to_idx[type_a]][cat_to_idx["Background"]] += 1
    
    # 3. SECOND PASS: Catch everything B found that A did not
    for idx_b, item_b in df_b.iterrows():
        if idx_b not in matched_in_b_indices:
            # Annotator B found it, A did not (Background row)
            type_b = item_b['Antipattern']
            conf_matrix[cat_to_idx["Background"]][cat_to_idx[type_b]] += 1

# --- 3. Kappa Calculation ---

N = np.sum(conf_matrix)

if N > 0:
    # Observed Agreement (Po)
    p_o = np.trace(conf_matrix) / N
    
    # Expected Agreement (Pe)
    row_sums = np.sum(conf_matrix, axis=1) # n_i.
    col_sums = np.sum(conf_matrix, axis=0) # n_.i
    
    p_e = np.sum((row_sums * col_sums) / (N**2))
    
    # Cohen's Kappa
    kappa = (p_o - p_e) / (1 - p_e) if p_e < 1 else 1.0
else:
    p_o, p_e, kappa = 0, 0, 0

# --- 4. Output Results ---

print(f"Total Unique Instances (N): {N}")
print(f"Observed Agreement (Po): {p_o:.4f}")
print(f"Expected Agreement (Pe): {p_e:.4f}")
print(f"Cohen's Kappa: {kappa:.4f}")

# Optional: View as DataFrame
conf_df = pd.DataFrame(conf_matrix, index=categories, columns=categories)
```
